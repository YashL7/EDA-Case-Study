```python
# Count plotting in logarithmic scale

def uniplot(df_appdata,col,title,hue =None):
    
    temp = pd.Series(data = hue)
    fig, ax = plt.subplots()
    width = len(df_appdata[col].unique()) + 7 + 4*len(temp.unique())
    fig.set_size_inches(width , 8)
    plt.xticks(rotation=45)
    plt.title(title)
    ax = sns.countplot(data = df_appdata, x= col, order=df_appdata[col].value_counts().index,hue = hue,palette='Wistia') 
        
    plt.show()
```

```python
# PLotting for income range

uniplot(target0_df,col='AMT_INCOME_RANGE',title='Distribution of income range',hue='CODE_GENDER')
```

Analysis of above graph:

-Female counts are higher than male for each income range bin.
-Count for income range 400000 and above is quite less.

```python
# Plotting for Income type

uniplot(target0_df,col='NAME_INCOME_TYPE',title='Distribution of Income type',hue='CODE_GENDER')
```

-For above listed income types, females have more credits than males.

```python
# Plotting for Contract type

uniplot(target0_df,col='NAME_CONTRACT_TYPE',title='Distribution of contract type',hue='CODE_GENDER')
```
